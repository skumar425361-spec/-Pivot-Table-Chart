**Easy Level**
1.  What is the purpose of a PivotTable in Excel?

**Answer:**   
A PivotTable is used to summarize, analyze, and organize large datasets quickly without changing the original data.
It helps users calculate totals, averages, and comparisons based on categories like region, product, or time.

2. Create a PivotTable to show Total Sales by Region.

**Answer:**  
* Select the complete dataset
* Go to insert → Pivot table
* Choose new worksheet
* Set fields as :-
  
    Row - Region

    Value - Sales (sum)

    [image](https://docs.google.com/document/d/11NJbRn7OHm1YhhbZI6WtRIoZhV2tZqymyCawC_FLXUs/edit?tab=t.0)

3. Apply Conditional Formatting to highlight sales greater than 400.

**Answer:**
* Select the Sales Column
* Go to Home →Conditional Formatting
* Choose Highlight Cell Rules → Greater Than
* Enter 400
* Click Ok

**Medium Level**

4. Explain the difference between Sorting and Filtering with an example from this dataset.

**Answer:**
1). **Sorting** arranges data in a specific order (ascending or descending ) based on one or more columns. It does not remove or hide any data, it only changes the order in which the data is displayed.  
**Example**: Arrange sales from highest to lowest

2). **Filtering** displays only the data that meets certain conditions and temporarily hides the rest of the data.

**Example:** Show only data for the east region.



5. Create a PivotChart showing Monthly Sales Trend.

**Answer:**  
* Select the Dataset
* Go to insert → Pivotable
* Choose Existing Worksheet and  Click Ok
* Set Fields as:

	Row → Date

	Value → Sales (sum)
* Right Click on any data in the pivot table
* Select Group
* Choose Months → Click Ok
* Click Anywhere inside the pivotable
* Go to insert → Pivotchart
* Select Line chart and Click Ok
[image](https://docs.google.com/document/d/1cPNRH8xzNetz9_KfWyMaGnOPptiRpSMPxLVLCax9U8c/edit?tab=t.0)

6. Use a Slicer to filter data by Product and observe changes in sales.

**Answer:**

* Go to insert →  Slicer
* Select Product

7. Why are Dashboards important for decision-making in business analysis?

# Product Filter (Slicer)

* Go to Insert → Slicer
